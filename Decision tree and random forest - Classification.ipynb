## Data preprocessing

```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing dataset
df=pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
#Checking size of dataset
df.shape
```

```python
#Checking values of purchased column
df.Purchased.value_counts()
```

```python
#Schematic information
df.info()
```

```python
#Descriptive statistics
df.describe()
```

```python
#Handling null values
df.isnull().sum()
```

```python
#Data visualization
sns.distplot(df["Age"])
plt.show()
```

```python
#Correlation data
df.corr()
```

```python
#Correlation heatmap
sns.heatmap(df.corr(),annot=True)
```

```python
#Checking for outliers
sns.boxplot(df.EstimatedSalary)
plt.show()
```

```python
#Checking for outliers
sns.boxplot(df.Age)
plt.show()
```

```python
#Splitting dependent and independent variables
#Independent variables
x=df.iloc[:,1:4]
x.head()
```

```python
#Dependent variable
y=df.Purchased
y.head()
```

```python
#Label encoding
from sklearn.preprocessing import LabelEncoder
```

```python
#Creating an object
le=LabelEncoder()
```

```python
#Encoding
x.Gender=le.fit_transform(x.Gender)
x.head()
```

```python
#Feature scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
#Creating object
ms=MinMaxScaler()
```

```python
#Scaling
x_scaled=pd.DataFrame(ms.fit_transform(x),columns=x.columns)
x_scaled.head()
```

```python
#Train-test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

### Model building - Decision tree 

```python
#Model building
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
pred=dtc.predict(x_test)
pred
```

```python
y_test
```

```python
df
```

```python
dtc.predict(ms.transform([[1,19,19000]]))
```

### Evaluation  

```python
#Evaluation of model
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
#Accuracy score
accuracy_score(y_test,pred)
```

```python
#Confusion matrix
confusion_matrix(y_test,pred)
```

```python
pd.crosstab(y_test,pred)
```

```python
#ROC-AUC curve
probability=dtc.predict_proba(x_test)[:,1]
probability
```

```python
#ROC curve
fpr,tpr,thresholds=roc_curve(y_test,probability)
```

```python
plt.plot(fpr,tpr)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```

```python
from sklearn import tree
plt.figure(figsize=(25,15))
tree.plot_tree(dtc,filled=True)
plt.show()
```

### Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
parameter={
    'criterion':['gini','entropy'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
grid_search=GridSearchCV(estimator=dtc,param_grid=parameter,cv=5,scoring="accuracy")
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
dtc_cv=DecisionTreeClassifier(criterion='entropy',
 max_depth= 4,
 max_features= 'sqrt',
 splitter= 'best')
dtc_cv.fit(x_train,y_train)
```

```python
pred=dtc_cv.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

### Model building - Random forest 

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
forest_params=[{'max_depth':list(range(10,15)),'max_features': list(range(0,14))}]
```

```python
rfc_cv=GridSearchCV(rfc,forest_params,cv=10,scoring="accuracy")
```

```python
rfc_cv.fit(x_train,y_train)
```

```python
pred=rfc_cv.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```
