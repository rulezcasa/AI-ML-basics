### Theoretical steps 

1. Choosing number of clusters (K)
2. Selecting at random K points, the centroids (not necessarily from the dataset)
3. Assign each data point to the closest centroid that from K clusters
4. Compute (Mean of cluster points) and place the new centroid of each cluster 
5. Reassign each data point to the new closest centroid. If any reassignment took place, go to step 4, otherwise the model is ready( convergence reached)

### Practical steps 

1. Import necessary libraries
2. Import dataset
3. Data preprocessing
4. Select the features to cluster
5. Find the optimal number of clusters -- elbow method
6. Train the model on the dataset using the optimal cluster x value
7. Visualize the clusters

### Importing libraries

### Selecting features to cluster
1. Annual Income
2. Spending Score

```python
x=df.iloc[:,3:].values
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

### Finding optimal number of clusters -- Elblow method

```python
from sklearn.cluster import KMeans
#initialise the list to store wcss values
wcss=[]
#Try different values of K (from 1 to 10) and calculate WCSS for each K
#K-Means__ to choose initial cluster centroids for k-means clustering
for k in range(1,11): 
    kmeans=KMeans(n_clusters=k,init="k-means++",random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
#Values of wcss from k=1 to k=10
wcss
```

```python
#plotting elbow curve
plt.figure(figsize=(8,6))
plt.plot(range(1,11),wcss,marker="o",color="red")
plt.title("Elbow curve")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.grid()
plt.show()
```

Based on the above curve, K is 5

### Training the model using optimal cluster k value

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=0)
#Return a label for data based on their cluster
y_kmeans=kmeans.fit_predict(x)
y_kmeans
```

### Visualize the cluster 

```python
#No. of datapoints belonging to the 0th cluster from 0th column annual income
x[y_kmeans==0,0]
```

```python
#PLotting cluster 1
# x axis -> annual income
# y axis -> spending score
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],color="red",label="cluster1")
plt.show()
```

```python
#PLotting all clusters
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],color="red",label="cluster1")
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],color="cyan",label="cluster1")
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],color="green",label="cluster1")
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],color="orange",label="cluster1")
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],color="blue",label="cluster1")
plt.xlabel("Annual income")
plt.ylabel("Spending score")
plt.show()
```

```python
#PLotting all clusters with centroids
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],color="red",label="cluster1")
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],color="cyan",label="cluster1")
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],color="green",label="cluster1")
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],color="orange",label="cluster1")
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],color="blue",label="cluster1")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color="yellow",s=300,label="centroid")
plt.title("K means clustering")
plt.xlabel("Annual income")
plt.ylabel("Spending score")
plt.legend()
plt.show()
```
