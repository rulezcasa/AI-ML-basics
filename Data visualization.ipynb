### Types of variables
1. Univariate 
2. Bivariate
3. Multivariate 

### Univariate
1. Dealing with one variable at a time 
2. Histrograms, bar charts, pie charts, boxplots 
3. Varaibles can be categorical or numerical

### Bivariate
1. Numerical vs Numerical -> Scatterplot
2. Numerical vs categorical -> ANOVA
3. Categorical vs Categorical > Chi squared error test

### Multivariate

1. Clustering analysis
2. Priniple component analysis
3. Multi correspondence analysis

#### Pie charts

```python
import matplotlib.pyplot as plt
#Plotting a pie chart
labels=["php","python","java","c"]
numbers=[75,70,80,60]
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis1.pie(numbers,labels=labels,autopct="%0.2f",colors=["orange","pink","green","blue"]) 
#autopct displays percent values
axis1.legend() #Displays a legend
```

```python
#To make a specific section stand out
plt.pie(numbers,labels=labels,explode=[0.1,0,0,0])
```

### Seaborn 

```python
#importing seaborn
import seaborn as sns
```

```python
#Datasets under seaborn library
print(sns.get_dataset_names())
```

```python
#Loading dataset
df=sns.load_dataset('tips')
df
```

```python
#Gives schematic of the dataset 
df.info()
```

```python
#To display the first n records
df.head(5)
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
```

```python
#Comparing size vs tips using scatterplot
sns.scatterplot(x="size",y="tip",data=df)
```

```python
#total_bill vs size using line plot
sns.lineplot(x="size",y="total_bill",data=df,ci=None) 
#ci removes confidence interval(standard deviation of the data)
```

```python
#Plotting a lined a histogram
sns.distplot(df["tip"])
```

```python
#Relation plot using day as hue
sns.relplot(x="total_bill",y="tip",data=df,hue="day")
```

```python
#Day vs tips using barplot (categorical variable vs numerical variable)
sns.barplot(data=df,x="day",y="tip")
```

```python
#Sex vs tips using barplot (categorical variable vs numerical variable)
sns.barplot(data=df,x="sex",y="tip",ci=None)
```

```python
#Sex vs tips using barplot with day as hue
sns.barplot(data=df,x="sex",y="tip",ci=None, hue="day")
```

```python
#Counting number of males and females using countplot
sns.countplot(x="sex",data=df)
```

```python
#Total bill vs tips using joinplot (Used for univaraite and bivaraite analysis)
sns.jointplot(x="total_bill",y="tip",data=df)

#the histogram represents univariate analysis
#the scatter plot represents bivaraite analysis
```

```python
#Boxplot for day vs total bill
sns.boxplot(x="day",y="total_bill",data=df)
```

Skewness

1. Negative -> More data towards Q3
2. Positive -> More data towards Q1
3. Symmetrical -> Equal data distribution
 

```python
#Correlation
corr=df.corr()
corr
#Values > 0.5 -> highly correlated
#Values < 0.5 -> less correlated
#Value = 1 -> perfect positive correlation
#Value = 0 -> No correlation
```

```python
#Plotting a heatmap to represent correlation
sns.heatmap(corr,annot=True,cmap="YlGnBu") #Cmap is used to change colors
```
